```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Datos de Facturacion Bueno.xlsx')
```

```python
df.info()
```

```python
df.head()
```

```python
filtro1=df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"]<=2000)]
filtro1
```

```python
filtro2=df[~(df["CVE_VEND"] == 5.0) & ~(df["CVE_VEND"] == 4.0)]
filtro2
```

```python
filtro3= df[df['FECHA_ENT']=="28/02/2022"]
filtro3
```

```python
filtro4= df[(df['CAN_TOT'] < 5951.7) | (df['STATUS']== "E")]
filtro4
```

```python
filtro5=df.loc[ : , ["CVE_DOC", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]
filtro5
```

```python
filtro6= df.iloc[[7001,7099], :]
filtro6
```
